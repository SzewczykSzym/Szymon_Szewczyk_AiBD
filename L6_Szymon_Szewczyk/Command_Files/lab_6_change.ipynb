```python
import pandas as pd

data = pd.read_csv('../Original_Data/14_WIELKOPOLSKIE.csv')

# some of the rows do not have the age of customer and the gender of a customer are not provided, so I remove them from the file

data_without_NaN = data.dropna(subset=['Wiek kupującego'])
data_without_NaN = data_without_NaN[data_without_NaN['Płeć kupującego'] != 'bd.'].reset_index(drop= True)
data_without_NaN = data_without_NaN.drop(columns=['Unnamed: 0'])

sorted_data = data_without_NaN.sort_values(['Marka', 'Ocena'])
sorted_data.to_csv(path_or_buf=f"../Analysis_Data/sorted_data.csv", index=False)
sorted_data
```
